### Take last four digits of CC Number

```python
def last_four(num):
    return str(num)[-4:]
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
df
```

```python
def yelp(bill):
    if bill < 10:
        return '$'
    elif bill >= 10 and bill < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['expns'] = df['total_bill'].apply(yelp)
```

```python
df
```

```python
df[(df['expns'] != '$') & (df['expns'] != '$$')]
```

### Using lambda expression

```python
df['total_bill'].apply(lambda bill: bill*2)
```

```python
df.head()
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
quality(10,1.01)
```

```python
df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
df[df['Quality'] != 'Other']
```
